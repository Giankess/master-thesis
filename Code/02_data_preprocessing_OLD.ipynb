Imports

```python
import pandas as pd
```

Combine Data

```python
covid_data = pd.read_csv('Data/fraser_covid_timeline.csv')
covid_data['Theme'] = 'Pandemic'
financial_crisis_data = pd.read_csv('Data/fraser_financial_crisis_timeline.csv')
financial_crisis_data['Theme'] = 'Financial Crisis'

full_data = pd.concat([covid_data, financial_crisis_data], ignore_index=True)
full_data.to_csv('Data/combined_timeline_data.csv', index=False)
```
